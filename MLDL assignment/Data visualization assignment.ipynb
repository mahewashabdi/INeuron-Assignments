```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
import seaborn as sns
```

```python
import pandas as pd
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
y=df['species']
```

```python
pca=decomposition.PCA(n_components=3)
```

```python
X_transformed=pca.fit_transform(df.drop(columns='species'))
```

```python
y
```

```python
color={'setosa':'red','versicolor':'blue','virginica':'green'}
```

```python
fig=plt.figure(figsize=(10,8))
ax=Axes3D(fig)
ax.scatter(X_transformed[:,0],X_transformed[:,1],X_transformed[:,2],c=y.map(color))
ax.set_title("3D plot of iris dataset")
ax.set_xlabel('1st dimension')
ax.set_ylabel('2nd dimension')
ax.set_zlabel('3rd dimension')
ax.legend(labels=y,loc='upper left')
plt.show()
```
